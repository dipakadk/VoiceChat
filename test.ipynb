```python
import requests
import json
from datetime import datetime, timedelta
import os
import time

# Configuration
API_URL = "https://api.mindbodyonline.com/public/v6/class/classschedules"
HEADERS = {
    "Api-Key": "1480df1197e94fecad6a23bb6136b3a4", 
    "SiteId": "5731506", 
}
OUTPUT_DIR = "./class_schedules"  # Directory to save the JSON files
DAYS_RANGE = 7  # Number of days in the future to fetch schedules
FETCH_INTERVAL = 15  # Interval to fetch data in seconds (24 hours = 86400 seconds)


def fetch_class_schedules(start_date: str, end_date: str) -> list:
    params = {
        "startDate": f"{start_date}T00:00:00",
        "endDate": f"{end_date}T23:59:59",
    }
    try:
        response = requests.get(API_URL, headers=HEADERS, params=params)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"API request failed: {e}")
        return []

    data = response.json()
    class_schedules = data.get("ClassSchedules", [])

    schedule_data_list = []
    for schedule in class_schedules:
        class_name = schedule.get("ClassDescription", {}).get("Name", "N/A")
        
        start_date = schedule.get("StartDate", "N/A")
        start_date_obj = datetime.fromisoformat(start_date)
        start_date_only = start_date_obj.date()

        end_date = schedule.get("EndDate", "N/A")
        end_date_obj = datetime.fromisoformat(end_date)
        end_date_only = end_date_obj.date()

        start_time = schedule.get("StartTime", "N/A")
        start_time_obj = datetime.fromisoformat(start_time)
        start_time_only = start_time_obj.time()

        end_time = schedule.get("EndTime", "N/A")
        end_time_obj = datetime.fromisoformat(end_time)
        end_time_only = end_time_obj.time()

        days = {
            "Sunday": schedule.get("DaySunday"),
            "Monday": schedule.get("DayMonday"),
            "Tuesday": schedule.get("DayTuesday"),
            "Wednesday": schedule.get("DayWednesday"),
            "Thursday": schedule.get("DayThursday"),
            "Friday": schedule.get("DayFriday"),
            "Saturday": schedule.get("DaySaturday")
        }

        true_days = [day for day, is_true in days.items() if is_true]

        schedule_data = {
            "ClassName": class_name,
            "StartDate": start_date_only.strftime('%Y-%m-%d'),
            "EndDate": end_date_only.strftime('%Y-%m-%d'),
            "AvailableDay": true_days[0] if true_days else "N/A",
            "StartTime": start_time_only.strftime('%H:%M:%S'),
            "EndTime": end_time_only.strftime('%H:%M:%S')
        }
        schedule_data_list.append(schedule_data)

    return schedule_data_list


def save_to_json(data: list, output_path: str):
    """
    Saves the data to a JSON file.
    """
    with open(output_path, "w") as json_file:
        json.dump(data, json_file, indent=4)
    print(f"Saved data to {output_path}")


def automate_class_schedule_fetch():
    """
    Automates fetching schedules, saving to JSON, and repeats daily.
    """
    while True:
        # Generate start and end dates dynamically
        today = datetime.today()
        start_date = today.strftime('%Y-%m-%d')
        print("The start date is ==================>",start_date)
        end_date = (today + timedelta(days=DAYS_RANGE)).strftime('%Y-%m-%d')
        print("The end date is==================>",end_date)

        print(f"Fetching schedules from {start_date} to {end_date}...")

        # Fetch schedules
        schedules = fetch_class_schedules(start_date, end_date)
        if not schedules:
            print("No schedules fetched. Retrying in the next interval.")
        else:
            # Ensure output directory exists
            os.makedirs(OUTPUT_DIR, exist_ok=True)

            # Save to a uniquely named JSON file
            output_file = os.path.join(
                OUTPUT_DIR,
                f"class_schedules_{start_date}_to_{end_date}.json"
            )
            save_to_json(schedules, output_file)

        print("Waiting for the next fetch interval...")
        time.sleep(FETCH_INTERVAL)  # Wait for 24 hours


if __name__ == "__main__":
    automate_class_schedule_fetch()
```
